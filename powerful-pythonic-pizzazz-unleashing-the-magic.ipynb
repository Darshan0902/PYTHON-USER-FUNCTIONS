  ## <span style="color: #FF4500;"> 5.) Printing all the rows :  </span>


```python
#Printing all rows in the dataset.

with open('/kaggle/input/500-albums-between-1955-2011/500 albums (1955-2011).csv','r') as csvfile:
    reader = csv.DictReader(csvfile)
# Using "for" loop to print each and every row.    
    for row in reader:
        print(row) 
```

  ## <span style="color: #FF4500;"> 6.) Printing first 100 rows :  </span>


```python
# First 100 rows of data :

with open('/kaggle/input/500-albums-between-1955-2011/500 albums (1955-2011).csv','r') as csvfile:
    reader = csv.DictReader(csvfile)
    
    rows = 101 #sing 101 rows to be stored in this above variable.
    for row in reader:
        rows -= 1 #Printing (101-1) rows or decreementing 101 by 1 such that we have now only 100 rows.
        
        if (rows>0): #If roes are greater than 0 they will be printed else not,
            print(row)
        else:
            break
```

  ## <span style="color: #FF4500;"> 7.) Shifting out data values into a list from a dictreader :  </span>


```python
# Copying data from a Dictreader to a list : 
with open("/kaggle/input/500-albums-between-1955-2011/500 albums (1955-2011).csv",'r') as csvfile:
    reader = csv.DictReader(csvfile)
    
    #Defining a empty list named albums.
    albums = []
    
    #Appending or adding the elements of file into the list.
    for row in reader:
        albums.append(row)
        

    
```

  ## <span style="color: #FF4500;"> 8.) Printing the number of records or rows :  </span>


```python
# Printing total number of rows or length of our list "albums".

print("Total numbers of Albums in our dataset :",len(albums))
```

  ## <span style="color: #FF4500;"> 9.) Using list comprehension instead of for loop  :  </span>
  #### <span style="color: orange;"> We will be printing the number of ablums that were released in the year 1974.</span>


```python
# Using List Comprehension to retrieve albums released in the year 1974 : 
albums_1974 = [row for row in albums if row["Year"] == "1974"]

# Checking the length of the albums released in year : 1974.
print("number of albums released in the year 1974:",len(albums_1974))
```

  ## <span style="color: #FF4500;"> 10.) Printing all the records in formatted manner:  </span>


```python
for albums in albums_1974:
    print(albums["Album"],"was released by",albums["Artist"])
    
# Using List compherension to print all the rows .
```
